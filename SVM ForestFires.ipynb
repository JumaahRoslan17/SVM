```python
import pandas as pd
import numpy as np
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
ff = pd.read_csv('forestfires (3).csv')
data = ff.describe()
```

```python
#drop month and days column
ff.drop(["month","day"],axis=1,inplace =True)
```

```python
predictors = ff.iloc[:,0:28]
target = ff.iloc[:,28]
```

```python
from re import X
def norm_func(i):
  X= (i-i.min())/(i.max()-i.min())
  return (X)
```

```python
fires = norm_func(predictors)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(predictors,target,test_size = 0.25, stratify = target)
```

```python
model_linear = SVC(kernel = "linear")
model_linear.fit(X_train,y_train)
pred_test_linear = model_linear.predict(X_test)
```

```python
np.mean(pred_test_linear ==y_test)
```

```python
#kernel = poly
model_poly = SVC(kernel= "poly")
model_poly.fit(X_train,y_train)
pred_test_poly = model_poly.predict(X_test)
```

```python
np.mean(pred_test_poly==y_test)
```

```python
#kernel = rbf
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(X_train,y_train)
pred_test_rbf = model_rbf.predict(X_test)
```

```python
np.mean(pred_test_rbf==y_test)
```

```python
#kernel = sigmoid
model_sig = SVC(kernel = "sigmoid")
model_sig.fit(X_train,y_train)
pred_test_sig = model_sig.predict(X_test)
```

```python
np.mean(pred_test_sig==y_test)
```
